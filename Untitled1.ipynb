```python
# Exploratory Data Analysis (EDA) on Cardiographic Dataset

# ================================
# 1. Import Libraries
# ================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# display settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# ================================
# 2. Load Dataset
# ================================
cardio_df = pd.read_csv("cardiographic.csv")

# Show first 5 rows
cardio_df.head()

# ================================
# 3. Data Cleaning and Preparation
# ================================
# Shape of dataset
print("Dataset shape:", cardio_df.shape)

# Info about dataset
display(cardio_df.info())

# Missing values check
print("Missing values per column:")
print(cardio_df.isnull().sum())

# Fill numeric missing values with column mean
num_cols = cardio_df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    cardio_df[col].fillna(cardio_df[col].mean(), inplace=True)

# Check categorical columns (if any)
cat_cols = cardio_df.select_dtypes(exclude=[np.number]).columns
for col in cat_cols:
    cardio_df[col].fillna(cardio_df[col].mode()[0], inplace=True)

# Detect outliers using IQR method
Q1 = cardio_df[num_cols].quantile(0.25)
Q3 = cardio_df[num_cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((cardio_df[num_cols] < (Q1 - 1.5 * IQR)) | (cardio_df[num_cols] > (Q3 + 1.5 * IQR))).sum()
print("Outliers per column:")
print(outliers)

# ================================
# 4. Statistical Summary
# ================================
summary = cardio_df.describe().T
summary["IQR"] = summary["75%"] - summary["25%"]
display(summary)

# ================================
# 5. Data Visualization
# ================================
# Histograms for numeric variables
cardio_df[num_cols].hist(figsize=(15, 10), bins=20)
plt.suptitle("Histograms of Numeric Features")
plt.show()

# Boxplots
plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(3, 4, i)
    sns.boxplot(y=cardio_df[col], color='skyblue')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cardio_df[num_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# Pairplot
sns.pairplot(cardio_df[num_cols].sample(200), diag_kind="kde")
plt.show()

# Violin plots for selected variables
plt.figure(figsize=(12, 6))
sns.violinplot(x="DL", y="LB", data=cardio_df, inner="quart", palette="Set2")
plt.title("Violin Plot: LB vs DL")
plt.show()

# ================================
# 6. Pattern Recognition and Insights
# ================================
# Example: Correlation of LB with other features
lb_corr = cardio_df.corr()["LB"].sort_values(ascending=False)
print("Correlation of LB with other variables:")
print(lb_corr)

# ================================
# 7. Conclusion (Markdown in Notebook)
# ================================
# In the final notebook, we will add markdown cells summarizing key insights,
# trends, and implications for decision-making.
```
