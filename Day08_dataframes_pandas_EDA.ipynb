## 8.3.2 accessing data from dataframe

```python
#dataframe[rows,columns]
#dataframe[startpos:endpos,startpos:endpos]
# .loc will be used to access the entire record of that perticular index
# .iloc will be used to access the columns with column and row index
```

```python
df.loc[20]
```

```python
df.loc[20,'Expenses'] # 20th index record and Expenses column value
```

```python
df['Univ'] # if we wanr to extract perticular column
```

```python
df.loc[:,'Univ'] ## all rows but Univ column
```

```python
df.loc[:2,'Univ'] 
```

```python
df.iloc[:2,1:4] # frist 2 rows and 3 columns frm indexing 1 to 3
```

```python
df.loc[[0,3],['SAT'	,'Top10','Accept']]
```
