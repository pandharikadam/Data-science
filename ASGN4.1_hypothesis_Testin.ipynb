```python
import math
from scipy.stats import norm

# Given values
sample_mean = 3050         # x̄
theoretical_mean = 4000    # μ
std_dev = 5 * 25           # σ = 125
n = 25                     # sample size

# Calculate standard error
standard_error = std_dev / math.sqrt(n)

# Calculate z-score
z_score = (sample_mean - theoretical_mean) / standard_error

# Calculate p-value (one-tailed, right side test: cost increased)
p_value = 1 - norm.cdf(z_score)

# Output results
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Weekly operating cost has significantly increased.")
else:
    print("Fail to reject the null hypothesis: No significant increase in weekly operating cost.")
```
