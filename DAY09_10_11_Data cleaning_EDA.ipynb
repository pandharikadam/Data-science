# 9 DAY 9_Data cleaning_EDA
WHat is EDA - EDA is nothing but extracting, cleaning and visualizing the data to identify patterns from the data.
In Python, EDA typically involves the following steps:
1) Loading the Data: Using libraries like pandas to load data into DataFrame structures.

2) Data Cleaning: Identifying missing values, outliers, or inconsistencies and handling them by filling, removing, or correcting the data.

3) Descriptive Statistics: Using methods like .describe() in pandas to compute basic summary statistics (mean, median, standard deviation, etc.) of the dataset.

4) Visualizing the Data: Using libraries such as matplotlib, seaborn, or plotly to create plots like histograms, box plots, scatter plots, and correlation heatmaps to visualize the distribution of data and relationships between variables.

5) Identifying Patterns: Looking for trends, patterns, and anomalies in the data that might suggest a deeper investigation or features for modeling.

### 9.5.3 Merge

```python
import numpy as np
d1 = {'movieid':[1,2,3,4,5],
     'title':['3 Idiots','2 States','Leo','RRR','Pushpa']}
movies = pd.DataFrame(d1)


d2 = {
    'memid':[1,2,3,4],
    'name':['Tom','Tim','Jon','Jim'],
    'movieid':[1,np.nan,3,5]
}
members=pd.DataFrame(d2)
```

```python
print(movies)
```

```python
print(members)
```

```python
pd.merge(movies,members) # it behaves like innner join. it it will find the same column name and join on the basis of that coulmn
```

```python
pd.merge(movies,members,on='movieid') # explicitly also we can mention the column name to join
```

```python
pd.merge(movies,members,on='movieid',how='right')
```

```python
pd.merge(movies,members,left_on='movieid',right_on='movieid',how='right') # if we have different column name
```

```python
pd.merge(movies,members,left_on='movieid',right_on='movieid',how='outer') 
```

```python
pd.merge(movies,members,left_on='movieid',right_on='movieid',how='outer') [['title','memid']]
```

#### 9.7.1 Variable unpacking
- Assigning values to the respective variable

```python
grp=df1.groupby('rank')# it will store all records based on rank values like for each catory of rank it will store all corresponding records with rank category
grp
```

```python
df1.head()
```

```python
for x in grp:# if we want to extract the data we need to iterate it. if want to do grouping based on single column
    print(x)# not alligned properly # after applying grouping only we can find out aggregated output
    print('_'*100)
```

```python
for x,y in grp:
    print(x)
    print('_'*100)

```

```python
for x,y in grp:
    print(y)
    print('_'*100) # If we see the difference between above 2 code. X - holds only group name, y- holds real records. It useful when we want to show aligned data
```

```python
grp.agg('mean',numeric_only='True')
```

```python
grp.agg('max',numeric_only=True)
```

```python
# grouping based on multiple column
grp2=df1.groupby(['rank','sex'])
for x,y in grp2:
    print(x,y)
    print('_'*50)
    
```

```python
grp2.agg('mean',numeric_only=True)
```

```python
# grp.agg(['mean','max','min'],numeric_only=True)# if we want to group based on multiple columns, we can use this
```

```python
grp3=df1[['service','salary','rank','sex']].groupby(['rank','sex']) # if we want to group based on multiple columns, extract all the coumns of no  data type along with aggregted column
for i in grp3:
    print(i)
    print('_'*50)
# grp3 -- u can check whether it has creaated or not
```

```python

grp3.agg(['min','max','mean'])
# E1 - rank and sex wise wise min,max,mean of phd,service and salary
```

```python
grp4=df1[['service','salary','rank','sex']].groupby(['rank','sex']) # if we want to group based on multiple columns(required for aggragtion), extract all the coumns along with aggregted column
# grp3 -- u can check whether it has creaated or not
grp4.agg(['min','max','mean'])
```
