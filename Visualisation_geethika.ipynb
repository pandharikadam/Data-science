#### Line plot

```python
# create a line plot
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.read_csv('../../datasets/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
series
```

```python
series.plot(figsize=(14,3), title='Daily Minimum Temperatures', xlabel='Time', ylabel='Minimum Temperature', color='cornflowerblue')
plt.show()
```

#### Histogram and Density Plots

```python
series.hist(figsize=(5,3),  color='#16a085', grid=False, edgecolor='white')
plt.show()
```

```python
# create a density plot
series.plot(kind='kde', color='purple', figsize=(5,3))
plt.show()
```

#### Box and Whisker Plots by Interval

```python
from pandas import Grouper
groups = series.groupby(Grouper(freq='YE'))
years = pd.DataFrame(index=range(365))
for name, group in groups:
    years[name.year] = group.values
years.boxplot(figsize=(7,3),grid=False)
plt.show()
```

#### Lag plot

```python
# create a scatter plot
from pandas.plotting import lag_plot
plt.figure(figsize=(6,3))
lag_plot(series, c='#76d7c4',edgecolor='midnightblue', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(6,3))
lag_plot(series, lag=30, c='#76d7c4',edgecolor='midnightblue', alpha=0.6)
plt.show()
```

```python
# create an autocorrelation plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(series,lags=90, 
         color="purple",
         title='Autocorrelation',
         auto_ylims=True,)
plt.show()
```

```python
plot_pacf(series,lags=90, 
         color="purple",
         title='Partial Autocorrelation',
         auto_ylims=True,)
plt.show()
```
