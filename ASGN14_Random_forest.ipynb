```python
1. Explain Bagging and Boosting methods. How is it different from each other.
 Bagging - 	Bagging builds multiple models with randomly selected data subsets, and it will aggregate output from all subset and decides an output.
	Initially it will selects the models with low BIAS and after aggregating algorithm makes high variance to low variance so eventually we will get low bias and low variance model.
	All models works parallel.
	Base model chosen as low bias and high variance.

2. Explain how to handle imbalance in the data.
- We can go for resampling the dataset or also we can go for Oversampling as well. 
```
