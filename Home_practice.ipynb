```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the population (uniform distribution)
population = np.array([1, 2, 3, 4, 5, 6])

# Step 2: Parameters for the experiment
sample_size = 3  # Size of each sample
num_samples = 1000  # Number of samples we will draw

# Step 3: Collect sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Step 4: Plotting the original population and the sample means
plt.figure(figsize=(12, 6))

# Plotting the population distribution
plt.subplot(1, 2, 1)
plt.hist(population, bins=6, alpha=0.7, color='skyblue', edgecolor='black')
plt.title("Population Distribution (Uniform)")
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plotting the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=20, alpha=0.7, color='lightgreen', edgecolor='black')
plt.title("Distribution of Sample Means")
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

# Show plots
plt.tight_layout()
plt.show()
```
