```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Classification

#### Read the data

```python
iris = pd.read_csv('../../datasets/iris.csv')
iris.head()
```

#### Encode the categorical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data = iris.copy()
```

```python
le = LabelEncoder()
#le.fit_transform(iris['species'])
```

```python
data['species'] = le.fit_transform(iris['species'])
data.head()
```

#### split the data

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.iloc[:,:4]
y = data.iloc[:,4]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

##### Build the model

```python
from sklearn.tree import DecisionTreeClassifier
```

#### C5.0 Algorithm

```python
dtree = DecisionTreeClassifier(max_depth=8, max_leaf_nodes=10, criterion='entropy', random_state=42)
dtree.fit(X_train,y_train)
```

##### Evaluate the model

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
yhat_train = dtree.predict(X_train)
cm = confusion_matrix(y_train,yhat_train)
print(cm)
print(classification_report(y_train,yhat_train))
```

```python
yhat_test = dtree.predict(X_test)
cm = confusion_matrix(y_test,yhat_test)
print(cm)
print(classification_report(y_test,yhat_test))
```

#### Visualise the tree

```python
from sklearn.tree import plot_tree
```

```python
cnames = iris['species'].unique()
cnames
```

```python
fnames = X.columns
fnames
```

```python
plt.figure(figsize=(10,10))
plot_tree(dtree,class_names=cnames, feature_names=fnames, filled=True)
plt.show()
```

#### Feature Importances

```python
dtree.feature_importances_
```

```python
X.columns
```

```python
fimp = pd.DataFrame(dtree.feature_importances_, index=X.columns, columns=['importance'])
fimp
```

```python
fimp.sort_values(by='importance').plot(kind='barh', title='Feature Importances', color='purple',figsize=(5,2), legend=False)
plt.show()
```

##### Visualise the data using the most important features

```python
sns.relplot(iris, x='petal_length', y='petal_width', hue='species', height=3,aspect=1.5)
plt.show()
```

#### Regression

```python
pdata = pd.read_csv('../../datasets/pollution.csv')
pdata.head(3)
```

#### Split the data

```python
X = pdata.iloc[:,1:]
y = pdata.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

##### Build the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
rmodel = DecisionTreeRegressor(max_depth=4)
rmodel.fit(X_train,y_train)
```

##### Evaluate the model

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
yhat_train = rmodel.predict(X_train)
tr_mse =  mean_squared_error(y_train,yhat_train)
tr_rmse = tr_mse**0.5
tr_r2 = r2_score(y_train,yhat_train)

np.round(tr_mse,2), np.round(tr_rmse,2),np.round(tr_r2,2)
```

```python
pdata['Pollution '].mean()
```

```python
yhat_test = rmodel.predict(X_test)
ts_mse =  mean_squared_error(y_test,yhat_test)
ts_rmse = ts_mse**0.5
ts_r2 = r2_score(y_test,yhat_test)

np.round(ts_mse,2), np.round(ts_rmse,2),np.round(ts_r2,2)
```
