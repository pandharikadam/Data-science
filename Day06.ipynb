#### 6.2 write a function GRADE which taken parameter and return the grades

```python
def grades(marks):
    if marks>90:
        print(f'obtained marks {marks} and garde is merit')
        return(print(f'obtained marks {marks} and garde is merit'))
    elif marks>80 and marks<90:
        print(f'obtained marks {marks} and garde is distinction')
        return(print(f'obtained marks {marks} and garde is merit'))
    else:
        print(f'obtained marks {marks} and garde is first class')
        return(print(f'obtained marks {marks} and garde is merit'))
```
