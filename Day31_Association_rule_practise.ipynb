```python
from mlxtend import frequent_patterns
from mlxtend.preprocessing import TransactionEncoder
```

```python
items_list=[['Milk','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
           ['Dill','Onion','Nutmeg','Kidney Beans','Eggs','Yogurt'],
            ['Milk','Apple','Kidney Beans','Eggs'],
            ['Milk','Unicorn','Corn','Kidney Beans','Yogurt'],
            ['Corn','Onion','Onion','Kidney Beans','Ice cream','Eggs']
           ]
```

```python
type(items_list)
```

```python
items_list
```

# Step 1 - Convert the transaction into 0,1 by using encoder

```python
enc=TransactionEncoder()
enc.fit_transform(items_list)
```

```python
# lets convert the above array output in dataframe for further use
```

```python
import pandas as pd
enc_encoded_df=pd.DataFrame(enc.fit_transform(items_list),columns=enc.columns_)
```

```python
enc_encoded_df.head()
```

# Step 2 - find out frequent itemset using apriori algorithm

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# To find out the frequent itemsest we need to mention the thereshold(min_support)
```

```python
frequent_items=apriori(enc_encoded_df,min_support=0.6,use_colnames=True)
frequent_items
```

# Step 3 - Find association rule

```python
assoc_rule=association_rules(frequent_items,min_threshold=1.0,num_itemsets=5)
# num_items - no of transactions
# confidance - min_threshold=1.0
# num_itemsets=5- no of transactions
assoc_rule
```

```python
assoc_rule[assoc_rule['lift']>1]
```

## 31.2) K-fold validation

```python
import pandas as pd
```

```python
data=pd.read_csv('pima_indians_diabetes.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
x=data.drop('class', axis=1)
y=data['class']
```

```python
x.head()
```

```python
y.head()
```

- First lets try the model validation by using train test split

#### a. Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# while evaluating any model use all theree methods. And compare the accuracy of alll three and decide models accuracy
1) confusion matrix, classification model,MSE,RMSE
2) K-fold
3) Leave one out
```

#### a.1) Build the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(max_iter=400)
```

```python
clf.fit(x_train,y_train)
```

#### a.2) predict - train data

```python
yhat_train=clf.predict(x_train)
```

```python
yhat_train
```

#### a.3) Evaluate the model - train data 

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
conf_mat=confusion_matrix(y_train,yhat_train)
print(conf_mat)
print(classification_report(y_train,yhat_train))
```

- accuracy - 0.77%

#### a.3) predict - test data 

```python
yhat_test=clf.predict(x_test)
```

#### a.4) Evaluate the model - test data 

```python
conf_mat_test=confusion_matrix(y_test,yhat_test)
print(conf_mat_test)
print(classification_report(y_test,yhat_test))
```

#### b. K-fold validation

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
kfold=KFold(n_splits=10)
```

```python
cvals=cross_val_score(clf,x,y,cv=kfold,scoring='accuracy')
cvals
```

- So it has given 10 accusracies based on splits. So we can consider avg of these 10 ouput would be our model accuracy

```python
cvals.mean()
```

- Model has 77% accuracy
