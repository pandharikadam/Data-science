# 47.1) Movie sentiment analysis

```python
import nltk
from nltk.corpus import stopwords
import pandas as pd
import re
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sent_ana=SentimentIntensityAnalyzer()
```

```python
a='This was a disgrace on the screen. An utter waste of time.'
```

```python
sent_ana.polarity_scores(a)
```

```python
b='Great movie. Thoroughly enjoyed it'
```

```python
sent_ana.polarity_scores(b)
```

```python
def get_vader_analysis(inputtext):
    sent_ana=SentimentIntensityAnalyzer()
    scores=sent_ana.polarity_scores(inputtext)
    if scores['compound']>0:
        return 'pos'
    else: 
        return 'neg'
```

```python
get_vader_analysis(a)
```

```python
get_vader_analysis(b)
```

# 47.2) Amazon review

```python
import numpy as np
import pandas as pd
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

```

```python
df = pd.read_csv('amazonreviews.tsv', sep='\t')
df.head()
```

```python
df['label'].value_counts()
```

```python
df.loc[0]['review']
```

```python
sid.polarity_scores(df.loc[0]['review'])
```

```python
get_vader_analysis(df.loc[0]['review'])
```

```python
df['comp_score'] = df['review'].apply(get_vader_analysis)
df
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(df['label'],df['comp_score'])
```

```python
print(classification_report(df['label'],df['comp_score']))
```

```python
print(confusion_matrix(df['label'],df['comp_score']))
```
