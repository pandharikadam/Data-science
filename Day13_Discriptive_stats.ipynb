## 13.1 Measures of frequency

```python
import pandas as pd

data = ['Apple', 'Banana', 'Apple', 'Orange', 'Banana', 'Apple', 'Mango', 'Banana', 'Apple', 'Orange']
df = pd.DataFrame(data, columns=['Fruit'])

# Frequency
freq = df['Fruit'].value_counts()

# Relative Frequency
rel_freq = df['Fruit'].value_counts(normalize=True) * 100

# Cumulative Frequency
cum_freq = df['Fruit'].value_counts().cumsum()

# Combine
summary = pd.DataFrame({
    'Frequency': freq,
    'Relative Frequency (%)': rel_freq.round(2),
    'Cumulative Frequency': cum_freq
})

print(summary)
```
