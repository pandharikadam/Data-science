```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
data=pd.read_csv('pima_indians_diabetes.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
# SPlit the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.iloc[:,0:8]
y=data.iloc[:,8]
```

```python
x.shape
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Build the model

```python
model=SVC(C=1,gamma=0.001)
# HIgh C - Leads to less margin, and may overfit
# Low C - Larger margin, better generalization
# Gamma = How much attention each support vector gets
# High gamma → Focuses only on nearby points
# Low gamma → Looks at a wider region
```

# fit the model

```python
model.fit(x_train,y_train)
```

# predict the model for train data

```python
y_hat_train=model.predict(x_train)
print(type(y_hat_train))
print(y_hat_train)
```

```python
cm=confusion_matrix(y_train,y_hat_train)
print(cm)
print(classification_report(y_train,y_hat_train))
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(cm,annot=True,cmap='viridis',fmt='d',cbar=False)
```

# predict the model for test data 

```python
y_hat_test=model.predict(x_test)
cm=confusion_matrix(y_test,y_hat_test)
print(cm)
print(classification_report(y_test,y_hat_test))
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(cm,annot=True)
```

- As test result and train results are not in same range, so model is overfitting

```python
param_id=[
          {'C':[0.6,1,1.5],
           'gamma':[0.002]
          }
         ]
```

```python
print(param_id)
print(type(param_id))
```

```python
model_hper=SVC()# initialise the model with empty paarmeter
gridsearch_model=GridSearchCV(model_hper,param_grid=param_id,cv=10)# initialise the model.CV- cross validation
gridsearch_model.fit(x_train,y_train)#fit the model with multiple parameters
```

```python
#print(grid_svc.best_estimator_,"\n")
#print(grid_svc.best_params_,"\n")
#print(grid_svc.best_score_,"\n")
```

```python
gridsearch_model.best_params_
```

# Hyperparameter model - predict on x train data

```python
yhat_hyper_train=gridsearch_model.predict(x_train)
cm_hyper_train=confusion_matrix(y_train,yhat_hyper_train)
print(cm_hyper_train)
print(classification_report(y_train,yhat_hyper_train))
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(cm_hyper_train,annot=True,fmt='d')
```

# Hyperparameter model - predict on x test data

```python
yhat_hyper_test=gridsearch_model.predict(x_test)
cm_hyper_test=confusion_matrix(y_test,yhat_hyper_test)
print(cm_hyper_test)
print(classification_report(y_test,yhat_hyper_test))
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(cm_hyper_test,annot=True)
```
