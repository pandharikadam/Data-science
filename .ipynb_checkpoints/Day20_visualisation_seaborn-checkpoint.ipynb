### 20.2.1) Bar plot

```python
sns.catplot(data=cars_df,x='Brand',y='Cars Listings',
           kind='bar',# type of chart
           height=3,
           aspect=2 # width=hight*aspect, so here it is 6
           )
plt.title('used cars listing')
plt.xticks(rotation=45)
plt.show()
```

### 20.2.1) Box plot

```python
tips_df.head()
```

```python
tips_df['day'].value_counts()
```

```python
sns.catplot(tips_df,x='day',y='total_bill',
           kind='box')
plt.show()
# In matplotlib boxplot takes one dimentional data
# in seaborn it takes 2 dimentional data but each group displays seperately, like below
```

### 20.2.3) stripp plot

```python
sns.catplot(tips_df,x='day',y='total_bill',
           kind='strip',
           palette='vlag')
plt.show()
```

### 20.2.4) sworm plot

```python
sns.catplot(tips_df,x='day',y='total_bill',
           kind='swarm',
           palette='vlag')
plt.show()
```

```python
sns.catplot(tips_df,x='day',y='total_bill',
           kind='box',
           palette='vlag',
           hue='sex')
plt.show()
```

```python
sns.catplot(tips_df,x='day',y='total_bill',
           kind='box',
           palette='vlag',
           hue='time')
plt.show()
```

## 20.2.2) Relplot

```python
mpg_df.head()
```

```python
sns.relplot(mpg_df,x='model_year',y='mpg',
          kind='line')
plt.show()
# shadow is nothing confidance interval. lineplot we will plot for where x and y has one to one relation. but here in one year many models might have released, so that shadow is there
```

```python
sns.relplot(mpg_df,x='model_year',y='mpg',
          kind='line',
            ci=False,
           hue='origin')
plt.show()
```

```python
sns.relplot(mpg_df,x='model_year',y='mpg',
          kind='line',
            ci=False,
           hue='origin',
           col='origin' # col - keeps each line in seperate plot
           )
plt.show()
```

```python
sns.relplot(tips_df,x='total_bill',y='tip',
              kind='scatter',
            hue='sex',
            size='size'
           )
plt.show()
```
