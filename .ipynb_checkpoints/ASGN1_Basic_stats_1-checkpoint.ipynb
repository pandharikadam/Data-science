## Step 2.4 - Encode Categorical Variables

```python
df_sales_data.head(3)
```

```python
df_sales_data1=df_sales_data.copy()
```

```python
# City - Dropping as it has only one distinct value
# SKU - as it is the idendifier to model
df_sales_data1.drop(columns=['SKU','City'],inplace=True)
```

```python
df_sales_data1.head()
```

### Step 2.4.1 Encoding for BU column

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
# Create encoder
encoder=OneHotEncoder()
# Fit and transform and convert into dataframe
BU_Encoded=pd.DataFrame(encoder.fit_transform(df_sales_data1[['BU']]).toarray(),columns=encoder.get_feature_names_out(['BU']))
```

```python
BU_Encoded.head()
```

```python
df_sales_data2=df_sales_data1.join(BU_Encoded)
```

```python
df_sales_data2.head()
```

### Step 2.4.2 Encoding for Day column

```python
Day_enoded=pd.DataFrame(encoder.fit_transform(df_sales_data1[['Day']]).toarray(),columns=encoder.get_feature_names_out(['Day']))
```

```python
df_sales_data2=df_sales_data2.join(Day_enoded)
```

```python
df_sales_data2.head()
```

### Step 2.4.3 Encoding for Brand column

```python
Brand_Encoded=pd.DataFrame(encoder.fit_transform(df_sales_data1[['Brand']]).toarray(),columns=encoder.get_feature_names_out(['Brand']))
#Day_enoded=pd.DataFrame(encoder.fit_transform(df_sales_data1[['Day']]).toarray(),columns=encoder.get_feature_names_out(['Day']))
```

```python
df_sales_data2=df_sales_data2.join(Brand_Encoded)
```

```python
df_sales_data2.head()
```

```python
df_sales_data2.shape
```

### Step 2.4.4 Encoding for Model column

```python
Model_encoded=pd.DataFrame(encoder.fit_transform(df_sales_data1[['Model']]).toarray(),columns=encoder.get_feature_names_out(['Model']))
Model_encoded.head()
```

```python
df_sales_data2=df_sales_data2.join(Model_encoded)
```

```python
df_sales_data2.shape
```

```python
df_sales_data2.head()
```

```python
df_sales_data2.columns
```

```python
df_sales_data2.drop(columns=['Day','BU','Brand','Model'],inplace=True)
```

```python
df_sales_data2.info()
```

```python
df_sales_data2.head()
```
