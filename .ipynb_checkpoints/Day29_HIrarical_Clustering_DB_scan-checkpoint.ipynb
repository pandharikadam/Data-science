```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Step 3 - Split the Data

# Step 4 – Scale the dataset – if required

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaled_data=scaler.fit_transform(udf)
```

```python
scaled_data
```

# Step 5 - Choose the Algorithm and Build the model

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='complete') # n_clusters - by default 2
model.fit(udf)
```

```python
# model
```

```python
# lets see the output. Any clustering modules gives label as the output
model.labels_
```

```python
from sklearn.metrics import silhouette_score

labels = model.labels_
score = silhouette_score(udf, labels)
print("Silhouette Score:", score)
```

```python
# lets attach label with main dataset to undertand the cluster
udf['label']=model.labels_
udf
```

```python
grp=udf.groupby('label')
grp.agg('mean')
 # or below code
# grp=udf.groupby('label').agg('mean')
# grp
```
