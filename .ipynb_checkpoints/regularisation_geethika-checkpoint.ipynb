```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the data

```python
rdata = pd.read_csv('../../datasets/boston.csv')
rdata.head(3)
```

```python
X = rdata.iloc[:,:13]
y = rdata.iloc[:,13]
```

#### 1. LinearRegression

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
```

```python
lr_model = LinearRegression()
lr_model.fit(X,y)
```

```python
lr_model.coef_
```

```python
model_params = pd.DataFrame()
model_params['features'] = X.columns
model_params['LReg'] = lr_model.coef_
model_params
```

##### Lasso Regression

```python
lasso = Lasso(alpha=1.0)
lasso.fit(X,y)
```

```python
lasso.coef_
```

```python
model_params['Lasso'] = lasso.coef_
model_params
```

#### Ridge Regression

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X,y)
```

```python
ridge.coef_
```

```python
model_params['Ridge'] = ridge.coef_
model_params
```

##### Elastic Net

```python
el_model = ElasticNet(l1_ratio=0.5, alpha=1.0)
el_model.fit(X,y)
```

```python
el_model.coef_
```

```python
model_params['ElastNet'] = el_model.coef_
model_params
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

sns.barplot(model_params, y='features', x='LReg',
            hue='features',
            palette='viridis',
            ax=axes[0,0])
axes[0,0].set_xlim(-20,5)
axes[0,0].set_title('Linear Regression Parameters')

sns.barplot(model_params, y='features', x='Lasso',
            hue='features',
            palette='viridis',
            ax=axes[0,1])
axes[0,1].set_xlim(-20,5)
axes[0,1].set_title('Lasso Regression Parameters')

sns.barplot(model_params, y='features', x='Ridge',
            hue='features',
            palette='viridis',
            ax=axes[1,0])
axes[1,0].set_xlim(-20,5)
axes[1,0].set_title('Ridge Regression Parameters')

sns.barplot(model_params, y='features', x='ElastNet',
            hue='features',
            palette='viridis',
            ax=axes[1,1])
axes[1,1].set_xlim(-20,5)
axes[1,1].set_title('ElasticNet Regression Parameters')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
