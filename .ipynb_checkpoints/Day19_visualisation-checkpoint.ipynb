```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Year=[1950,1960,1970,1980,1990,2000,2010,2020]
gdp=[300,540,670,980,1020,1100,1300,1470]
```

```python
# line chart
plt.plot(Year,gdp)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot(Year,gdp,label='GDP') # here it label='GDP' doesnot make any sense as we have only one columns to compare. it helps to display legend
plt.title('GDP over the years')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend()
plt.show()
# plt.plot - plots the ine plot
```

```python
plt.figure(figsize=(6,3))
plt.plot(Year,gdp,label='gdp',linestyle='dotted')#solid,dashed,dotted
plt.title('gdp over the years')
plt.xlabel('Years')
plt.ylabel('gdp')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.plot(Year,gdp,
         label='gdp',
         linestyle='dashed',
        marker='H',
        markersize=10,
        markerfacecolor='#900c3f',markeredgecolor='grey')
plt.title('gdp over the years')
plt.xlabel('Years')
plt.ylabel('gdp')
plt.legend()
plt.show()
```

#### 19.1.2 Nr vs Nr - scatter plot(numeric vs numeric)

```python
hprice=pd.read_csv('kc_house_data.csv')
hprice.head()
```

```python
# lets take only first 100 records
data=hprice.head(100)
plt.scatter(x=data['sqft_living'],
            y=data['price'],
            s=50,#size of dots
            c='orange',
            alpha=0.6 # transperancy
           )
plt.title('sqft vs house price')
plt.xlabel('sqft')
plt.ylabel('house price')
plt.show
```

```python
plt.plot(hprice['sqft_living'],hprice['price'])
plt.show()
```

```python
plt.scatter(x=hprice['sqft_living'],y=hprice['price'])
plt.show()
```

```python
data=hprice.head(100)
plt.scatter(x=data['sqft_living'],
            y=data['price'],
           s=data['sqft_lot']/100,#size of dots/marker size
           c=data['floors'],# colours of dots
            alpha=0.6 # transperancy
           )
plt.title('sqft vs house price')
plt.xlabel('sqft')
plt.ylabel('house price')
plt.show
```

#### 19.1.2 categorical vs numeric

##### 19.1.2.1 Bar chart

```python
course=['datascience','analytics','tab','pb','ms']
students=[78,120,28,88,40]
```

```python
plt.figure(figure=(7,4))
plt.bar(course,students,
            edgecolor='black',
            alpha=0.8)
plt.title('couses vs students')
plt.xlabel('course')
plt.ylabel('students')
plt.show()
```

##### 19.1.2.2 pie chart

```python
plt.pie(students,labels=course,shadow=True)
plt.show()
```

```python
plt.pie(students,labels=course,shadow=True,
        autopct='%0.0f%%',
       wedgeprops=dict(edgecolor='grey',linewidth=1),
       explode=[0,0,0.2,0,0])
plt.show()
```
