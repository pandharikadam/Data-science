```python
import pandas as pd
```

##### Read the data

```python
iris = pd.read_csv('../../datasets/iris.csv')
iris.head()
```

### 1. Encode Categorical Data

##### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
iris_label = iris.copy()
```

```python
iris_label.head(3)
```

```python
le = LabelEncoder()
le.fit_transform(iris_label['species'])
```

```python
iris_label['species'] = le.fit_transform(iris_label['species'])
iris_label.head(3)
```

##### One Hot Encoding

###### a. OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
iris_ohe = iris.copy()
iris_ohe.head(3)
```

```python
ohe_enc = OneHotEncoder()
y_enc = pd.DataFrame(ohe_enc.fit_transform(iris_ohe[['species']]).toarray())
y_enc
```

```python
iris_new = iris_ohe.join(y_enc)
iris_new.drop('species', axis=1, inplace=True)
iris_new
```

```python
iris_new.rename(columns={0:'species_setosa', 1:'species_versicolor', 2:'species_virginica'}, inplace=True)
iris_new
```

### 3. Isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination=0.02)
iso.fit(iris_label)
```

```python
iso.predict(iris_label)
```

```python
iris_label.shape
```

```python
iris_label['anomaly'] = iso.predict(iris_label)
iris_label
```

```python
iris_label[iris_label['anomaly'] == -1]
```
