```python
Background
In quality control processes, especially when dealing with high-value items, 
destructive sampling is a necessary but costly method to ensure product quality. 

The test to determine whether an item meets the quality standards destroys the item, 
leading to the requirement of small sample sizes due to cost constraints.
    
Scenario
A manufacturer of print-heads for personal computers is interested in estimating the mean durability of their 
print-heads in terms of the number of characters printed before failure. To assess this, 
the manufacturer conducts a study on a small sample of print-heads due to the destructive nature of the testing process.

Data
A total of 15 print-heads were randomly selected and tested until failure. 
The durability of each print-head (in millions of characters) was recorded as follows:
1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29

Assignment Tasks
a. Build 99% Confidence Interval Using Sample Standard Deviation
Assuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters 
printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using 
the t-distribution for this task.
                  
b. Build 99% Confidence Interval Using Known Population Standard Deviation
If it were known that the population standard deviation is 0.2 million characters, construct a 99% confidence interval for 
the mean number of characters printed before failure.
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stat
```

- Below are the  given values in above problem statement
- c=95(default)
- sample vaule(n=15)
- Degree fo freedom = n-1 = 15-1=14
- As the given sample values are n<30, we will be using t distribution to find out the confidance interval

```python
printer_head_char=pd.DataFrame([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29],columns=['no_of_char'])
```

```python
printer_head_char.head(3)
```

```python
printer_head_char.describe()
```

```python
sample_mean=np.round(printer_head_char['no_of_char'].mean(),3)
print(sample_mean)
```

```python
sample_stddev=np.round(printer_head_char['no_of_char'].std(),3)
print(sample_stddev)
```

# a. Build 99% Confidence Interval Using Sample Standard Deviation

```python
conf_int_ttest=stat.t.interval(0.99,df=14,loc=1.239,scale=0.193/np.sqrt(15))
np.round(conf_int_ttest,3)
# 0.19 - sample stddev
```

- so, on an avg printer will prints the character between 1.09M to 1.39M 

# b. Build 99% Confidence Interval Using Known Population Standard Deviation

- Population standard deviation is given 0.2, so we need to use t test

```python
conf_int_ztest=stat.norm.interval(0.99,loc=1.239,scale=0.2/np.sqrt(15))
np.round(conf_int_ztest,3)
```

- so, on an avg printer will prints the character between 1.09M to 1.39M
