```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series=pd.read_csv('daily-minimum-temperatures.csv',index_col=0,header=0,parse_dates=True)
```

```python
series.head()
```

```python
series['Temp'].max()
```

```python
series['Temp'].idxmax()
```

```python
series['Temp'].idxmin()
```

```python
series.info()
```

```python
series.plot(figsize=(14,3),
                   xlabel='TIme',
                   ylabel='minimum temp')
plt.show()
```

```python
series.hist(grid=False,edgecolor='white')
plt.show()
```

```python
series.plot(kind='kde')
```

```python
from pandas import Grouper
groups = series.groupby(Grouper(freq='YE'))
years = pd.DataFrame(index=range(365))
for name, group in groups:
    years[name.year] = group.values
years.boxplot(figsize=(7,3),grid=False)
plt.show()
```

- LAG plot

```python
from pandas.plotting import lag_plot
plt.figure(figsize=(6,3))
lag_plot(series,edgecolor='midnightblue', alpha=0.6)
plt.show()
```

```python
from pandas.plotting import lag_plot
plt.figure(figsize=(6,3))
lag_plot(series,edgecolor='midnightblue', alpha=0.6,lag=30)
plt.show()
```

```python
from pandas.plotting import lag_plot
plt.figure(figsize=(6,3))
lag_plot(series,edgecolor='midnightblue', alpha=0.6,lag=60)
plt.show()
```

- Atutocorellation

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(series,lags=90, 
         color="purple",
         title='Autocorrelation',
         auto_ylims=True,)
plt.show()
```

```python
plot_pacf(series,lags=90, 
         color="purple",
         title='Partial Autocorrelation',
         auto_ylims=True,)
plt.show()
```
