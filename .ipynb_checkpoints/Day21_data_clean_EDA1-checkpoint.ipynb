###  Step 2.6.1 - Outlier Detection - ozone

```python
sns.distplot(data['Ozone'])
plt.show()
```

- as the data distribution is skewed we need to use interquartile method to remove the outlier

```python
# lets cap the outliers to upper limit
```

```python
Q1 = data['Ozone'].quantile(0.25)
Q3 = data['Ozone'].quantile(0.75)
IQR = Q3 - Q1
Q1, Q3, IQR
```

```python
upper_limit = Q3 + 1.5 * IQR
upper_limit
```

```python
data[data['Ozone']>upper_limit]
```

```python
data['Ozone']=np.where(data['Ozone']>upper_limit,upper_limit,data['Ozone'])
```

```python
data[data['Ozone']>upper_limit]
```

```python
###  Step 2.6.2 - Outlier Detection - Solar
```

```python
sns.distplot(data['Solar'])
plt.show()
```

- as the data distribution is skewed we need to use interquartile method to remove the outlier

```python
Q1_solar = data['Solar'].quantile(0.25)
Q3_solar = data['Solar'].quantile(0.75)
IQR_solar = Q3_solar - Q1_solar
Q1_solar, Q3_solar, IQR_solar
```

```python
upper_limit_solar = Q3_solar + 1.5 * IQR_solar
upper_limit_solar
```

```python
data['Solar']=np.where(data['Solar']>upper_limit_solar,upper_limit_solar,data['Solar'])
```

```python
sns.distplot(data['Solar'])
plt.show()
```

```python
sns.catplot(data,y='Solar',kind='box')
plt.title('Solar')
plt.show()
```
