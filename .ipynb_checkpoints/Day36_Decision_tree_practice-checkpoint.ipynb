## Encode the species column as it is in cahracter format

```python
from sklearn.preprocessing import LabelEncoder
```

```python
iris_1=iris.copy()
```

```python
label_enc=LabelEncoder()
label_enc.fit_transform(iris_1['species'])
```

```python
# attach these values as column
```

```python
iris_1['species']=label_enc.fit_transform(iris_1['species'])
```

```python
iris_1.head()
```

```python
## Split the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=iris_1.iloc[:,:4]
y=iris_1.iloc[:,4]
```

```python
x.shape
```

```python
y.shape
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

## build the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier(max_depth=8,max_leaf_nodes=10,random_state=42,criterion='gini') # criterion='gini' by default we can {"gini", "entropy", "log_loss"} there is no difference in output
dtree.fit(x_train,y_train)
# when we are using criterion='gini' then the algorithm is called as CART
# when we are using criterion='entropy' then the algorithm is called as C5.o
```

## predict

```python
yhat_train=dtree.predict(x_train)
#yhat_train.head()
yhat_train
```

## evaluate the model

```python
# as it is classification matrix lets use metric like consuion matrix, classfication report
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_train,yhat_train)
print(cm)
class_rep=classification_report(y_train,yhat_train)
print(class_rep)
```

- 100% accuracy

## lets check whether it is overfitting/underfitting or not
- To do this need to evaluate the model with x_Test data

```python
yhat_test=dtree.predict(x_test)
#yhat_train.head()
yhat_test
```

```python
cm_test=confusion_matrix(y_test,yhat_test)
print(cm_test)
class_rep_test=classification_report(y_test,yhat_test)
print(class_rep)
```

- no overfitting or underfitting as train dataset accuracy and test dataset accuracy almost in same range

## Visualise the tree

```python
from sklearn.tree import plot_tree
```

```python
class_names=iris['species'].unique()
class_names
```

```python
feature_names=x.columns
feature_names
```

```python
plt.figure(figsize=(15,15))
plot_tree(dtree,feature_names=feature_names,class_names=class_names,filled=True)
plt.show()
```

```python
# we can find whicha are the feature imporatnant for model.
dtree.feature_importances_
```

```python
x.columns
```

```python
Imp_feature=pd.DataFrame(dtree.feature_importances_,index=x.columns)
Imp_feature
```

```python
Imp_feature.plot(kind='barh',title='Imp_feature')
plt.show
```
