# DAY_7 CLASSES AND OBJECTS AND MODULES
- Classes - are not real
- In Python, a class is a blueprint for creating objects (instances). It defines a set of attributes (variables) and methods (functions) that the created     objects will have. A class allows you to group related data and functions together in a logical structure.
- Clsses are also a data type
- Objects - are real and created based on templates
- Module - is a collectin of classes or functions

```python
class person:
    def __init__(self,name,age,gender):# __init__ constructor we can deiffine methods(self,name,age,gender)
        self.name=name
        self.age=age
        self.gender=gender
    # Lets define the functionality of each method
    def smile(self): # smile is method and methods are bound functions with objects. This is a kind of function, in which u can write any logic by using parameters passed to class
        print(f'{self.name} is smiling')

    def talk(self):
        print(f'{self.name} is talking')

    def walk(self):
        print(f'{self.name} is walking')

    
        
```

''' create a class of student which can take name and grade as an input. 
    and also create the methods
        1) student_info which shows grades and name
        2) shows_merit which shows firstclass,secondclass or ditinction '''

```python
class student:
    def __init__(self,name,grade):
        self.name=name
        self.grade=grade
    def student_info(self):
        print(f"the student : {self.name}, has garde : {self.grade}")

    def shows_merit(self):
        if self.grade>60:
            print(f"the student : {self.name}, is passed in first garde with : {self.grade}")
        else:
            print(f"the student : {self.name}, is passed in second grade with : {self.grade}")
                

```

```python
# step 1 - create an object
student_Details=student('Ram',50)
# step 2 - call required method
student_Details.student_info()
```

```python
student_Details.shows_merit()
```
