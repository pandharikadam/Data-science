```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#### Rea the data

```python
data = pd.read_csv('../../datasets/Universities.csv', index_col='Univ')
data.head()
```

##### Standardise the data ( mandatory step for PCA)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
#scaled_data
```

##### Reduce the dimensionality

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=6)
pdata = model.fit_transform(scaled_data)
#pdata
```

```python
ratio = model.explained_variance_ratio_
ratio
```

```python
vars = np.cumsum(ratio)
vars
```

#### Data to build model retaining 90% of information - i.e include 2 Principal Components

```python
model_data = pd.DataFrame(pdata[:,:2], columns=['PC1','PC2'])
model_data.head()
```

##### Visualise the data

```python
plt.figure(figsize=(6,3))
plt.plot(range(1,7),vars, linestyle='dotted', marker='H', markerfacecolor='orange')
plt.title('Explained Variances')
plt.xlabel('Principal Components')
plt.ylabel('Information Retained' )
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(x=range(1,7),height=ratio, color='purple')
plt.title('Explained Variances')
plt.xlabel('Principal Components')
plt.ylabel('Information Retained' )
plt.show()
```

### TSNE - For Visualisation

```python
from sklearn.manifold import TSNE
```

```python
iris = pd.read_csv('../../datasets/iris.csv')
iris.head()
```

```python
features = iris.iloc[:,:4]
features.head()
```

```python
tsne = TSNE(n_components=2)
tdata = tsne.fit_transform(features)
#tdata
```

```python
tdf = pd.DataFrame(tdata, columns=['c1','c2'])
tdf['species'] = iris['species']
tdf.head()
```

```python
sns.relplot(tdf, x='c1', y='c2', kind = 'scatter', hue='species', height=3, aspect=1.5)
plt.show()
```
