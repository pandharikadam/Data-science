## Accessing value sfrom dict 

```python
states=['MH','KA','GJ','MP','AP']
capital=['mumbai','bang','ahemadabad','patna','hyderabad']
zip(states,capital)
dstates = dict(zip(states,capital))
print(dstates)
```

```python
dstates['MH']
```

```python
dstates['UP']
```

### .get

Return the value for key if key is in the dictionary, else default.

```python
dstates.get('MH')
```

```python
dstates.get('UP')
```

```python
dstates.get('UP','Not found')
```

```python
dstates['UP']='lucknow'## If mentioned index is there it will modify, if it is not there it will add
```

```python
dstates
```

### .setdefault
Signature: dstates.setdefault(key, default=None, /)
Docstring:
Insert key with a value of default if key is not in the dictionary.

Return the value for key if key is in the dictionary, else default.

```python
dstates.setdefault('MH','pune') # if spacific index available it will not modify the value
```

```python
dstates.setdefault('UP','fgua')
```

```python
dstates.setdefault('KR','kochi') # # if spacific index is not available it will add key value pair to dict
```

```python
dstates
```

```python
d1={1:'a',2:'b',3:'c'}
d2={1:'e',2:'f',4:'c'}
d1.update(d2) # update function overrides if it mattching the index, if it is not matcing it will add that key:pair to value to first dict 
```
