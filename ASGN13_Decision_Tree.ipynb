```python
1. What are some common hyperparameters of decision tree models, and how do they affect the model's performance? 
-- max_depth - Small depth → simpler model → lower variance, higher bias (underfitting)
                Large depth → complex model → low bias, high variance (overfitting)

-- max_leaf_nodes - Directly limits model complexity
                    Smaller value → more regularized tree

```

```python
2. What is the difference between the Label encoding and One-hot encoding?
-- Label encoding - will assign unique value to a specific category.
                    - important when we have ordinal importance.

-- one hot encoding - will assign a seperate column to a specific category.
                    - important when we dont have ordinal importance.
```
