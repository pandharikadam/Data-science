## 4.1 SET
- Set is collectin of unique elments in python.
- Sets do not allow duplicate values.
- Sets do not maintain any perticular order, so we can not apply any indexing.
- Sets are mutable, but individual element of sets can not be replace/modified.

### 4.1.1 There are two ways to create set
1) normal war - my_set={1,2,3,4,5}
2) by using set constructor - my_Set=set([1,2,3,4,5])


### 4.1.2 Set operations/methods
- add() - it will add new element to set, it can add anywhere.
- remove() -  it will remove the element from the set. If given element is available in set than it will remove or else it will throgh an error.
- discard() - it will remove the element from the set. If given element is available in set than it will remove or else it will not throgh an error.
- .union - it displays all the elements from both the set and removes the duplicate
- .interset - commomn elements from both the set
- .difference - unique elements frrom set 1
- .symmentric_difference - only unique elemtns from both sets

```python
# Create 2 sets and try all four set operations
set1={1,2,3,4}
set2={3,4,5,6}
union_set=set1.union(set2)
print(union_set)
intset=set1.intersection(set2)
print(intset)
diff_set=set1.difference(set2)
print(diff_set)
symmetric_diff_set=set1.symmetric_difference(set2)
print(symmetric_diff_set)
```
