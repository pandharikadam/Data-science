```python
iris_features=iris.iloc[:,0:4]
iris_features.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
model=DBSCAN(min_samples=12, eps=0.5) # min_samples thumb rule is decide this value is take twice or thrice of your no of features
model.fit(iris_features)
```
