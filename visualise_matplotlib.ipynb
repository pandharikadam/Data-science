```python
year = [1950,1960,1970,1980,1990,2000,2010,2020]
gdp = [300,540,670,980,1020,1100,1300,1470]
```

```python
plt.figure(figsize=(6,3))
plt.title('GDP over the Years')
plt.xlabel('Year')
plt.ylabel('GDP')

# Line chart
plt.plot(year,gdp, 
         label='GDP',
         linestyle='dotted', linewidth=1,
         marker='H', markersize=8, markerfacecolor='#900c3f', markeredgecolor='grey'
        )

plt.legend()
plt.show()
```

#### b. Numeric vs Numeric - Scatter Plot

```python
hprice = pd.read_csv('datasets/kc_house_data.csv')
hprice.head()
```

```python
data = hprice.head(100)
plt.figure(figsize=(6,3))
plt.title('King County House prices', fontsize=14)
plt.xlabel('Living Area in SqFt.', color='midnightblue')
plt.ylabel('House Price (in Millions)',color='midnightblue')
plt.scatter(x=data['sqft_living'], 
            y=data['price']/100000,
            s= 50,
            c='orange',
            edgecolor='midnightblue',
            alpha=0.7
           )
plt.show()
```

```python
data = hprice.head(100)
plt.figure(figsize=(6,3))

plt.title('King County House prices', fontsize=13)
plt.xlabel('Living Area in SqFt.', color='midnightblue')
plt.ylabel('House Price (in Millions)',color='midnightblue')

plt.scatter(x=data['sqft_living'], 
            y=data['price']/100000,
            s= data['sqft_lot']/100,
            c=data['floors'],
            edgecolor='midnightblue',
            alpha=0.6
           )
plt.show()
```

#### a. Categorical vs Numeric - Bar Chart

```python
courses = ['Data Science','Analytics','Tableau','PowerBI','MySql']
students = [78,120,26,88,40]
```

```python
plt.figure(figsize=(7,3))

plt.title('Students Enrolled', fontsize=13)
plt.xlabel('Course')
plt.ylabel('No. of Students')

x = plt.bar(courses, students,
        color='cornflowerblue', edgecolor='black',
        alpha=0.6,
        width=0.6       
       )
plt.xticks(fontsize=8,
           #rotation=90
          )
plt.yticks(fontsize=8)
plt.bar_label(x)

plt.show()
```

```python
plt.figure(figsize=(7,3))

plt.title('Students Enrolled', fontsize=13)
plt.ylabel('Course')
plt.xlabel('No. of Students')
# Horizontal Bar
x = plt.barh(courses, students,
        color='cornflowerblue', edgecolor='black',
        alpha=0.6,
        #height=0.6       
       )
plt.xticks(fontsize=8,
           #rotation=90
          )
plt.yticks(fontsize=8)
plt.bar_label(x)

plt.show()
```

#### b. Categorical vs Numeric - Pie Chart

```python
courses = ['Data Science','Analytics','Tableau','PowerBI','MySql']
students = [78,120,26,88,40]
```

```python
plt.title('Students Enrolled', fontsize=13)

plt.pie(students, labels=courses,
        autopct ='%0.0f%%',
        textprops = dict(fontsize=8, color='black', 
                         #rotation=45
                        ),
        colors = ['tomato','cornflowerblue','cyan','pink','#cacfd2'],
        wedgeprops = dict(edgecolor='grey',linewidth=1,linestyle='solid'),
        explode = [0,0,0.2,0,0]
       )
plt.show()
```
