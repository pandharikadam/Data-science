## 18.1 One sample Z-test

#### Ex 18.1.1 Testing if new diet changes the averege weight of individuals 
A nutrition claims that average wight of individuals on a specific diet is 70kg with a standard deviation of 3.5. A sample of 40 people following the diet is tested and their avg wight is mesured as 67.9. Test id the nutrionists claim is true.

```python
# As it is talking about wight(number) and sample size is more than 30 and the n of samples are only one.
# So the test conducted would be "1 sample z test"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
#ttest
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel
#ztest
#import statsmodels.api as sn



# Step1 - set up the null hypothesis
H0='Avg weight of individual is 70kg'    # H0:Avg weight of individual is 70kg - H0:meaw=70
Ha='Avg weight of individual is not equal to 70kg'    # Ha:Avg weight of individual is not equal to 70kg - Ha:meaw<>70
# As meaw<>70 so it is a 2 tail test

# Step2 - Set the threshold
    # As alpha is not given then we can take default alpha
alpha=0.05

# step 3 - Colllect the evidences
population_mean=70 # meaw
pop_std_dev=3.5 #sigma_std_dev
sample_size=40 #n
sample_mean=67.9

# Step 4 - find p and z value
# Z-test

Z_score_weight = (sample_mean-population_mean)/(pop_std_dev/np.sqrt(sample_size))
    # We dont have z-test readily  availabe so we have built our own procees to conduct z-test
print(f" zscore : {Z_score_weight:,.3f}")
    #round(Z_score_weight,2)
p_value_weight = 2*stats.norm.cdf(Z_score_weight) # as it is a 2 tail test, always we compare p=2alpha, so multiplied by 2 in formula
print(f" pvalue : {p_value_weight}")

# Step 5 - conclusion compare p and alpa
if p_value_weight<alpha:
    print(f' reject the null hypothesis : after new diet : {Ha} : diet changes the averege weight of individuals')
else:
    print(f' fail to reject null hypothesis : after new diet : {H0} : diet not changes the averege weight of individuals')
```

A factory claims that its lights bulbs lasts on average 800 hours with a std dev of 20. A random sample of 50 light bulbs is tested to see if they last longer. The sa,ple mean is 811.97

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
#ttest
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel
#ztest
import statsmodels.api as sn



# Step1 - set up the null hypothesis
H0='factory lifght bulbs last on average of 800 hours'    # H0: H0:meaw<=800
Ha='factory lifght bulbs lasts more than 800 hours'    # Ha: Ha:meaw>800
# As meaw>70 so it is a 1 tail(right) test

# Step2 - Set the threshold
    # As alpha is not given then we can take default alpha
alpha=0.05

# step 3 - Colllect the evidences
population_mean=800 # meaw
pop_std_dev=20 #sigma_std_dev
sample_size=50 #n
sample_mean=811.97

# Step 4 - find p and z value
# Z-test

Z_score_weight = (sample_mean-population_mean)/(pop_std_dev/np.sqrt(sample_size))
    # We dont have z-test readily  availabe so we have built our own procees to conduct z-test
print(f" zscore : {Z_score_weight:,.3f}")
    #round(Z_score_weight,2)
p_value_weight = 1-stats.norm.cdf(Z_score_weight) # as it is a 1 tail(right tail) test, always we compare p=alpha, so the formula 1-stats.norm.cdf(Z_score_weight)
print(f" pvalue : {p_value_weight}")

# Step 5 - conclusion compare p and alpa
if p_value_weight<alpha:
    print(f' reject the null hypothesis : factory light bulbs last : {Ha} ')
else:
    print(f' fail to reject null hypothesis : factory light bulbs last : {H0} ')
```

## 18.2 one sample t-test

#### Ex- 18.2.1 : Average exam score vs. National exam score
A university claims that students score 75% on an average in national statistics exam. 
collected a random sample of 20 students and test if their mean is significanlty different.

```python
sample_Scores=[75,79,85,76,56,61,80,69,65,61,77,65,90,69,72,70,63,74,76,66]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
#ttest
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel
#ztest
import statsmodels.api as sn



# Step1 - set up the null hypothesis
H0='sample mean same as population mean , tno significant score difference'    # H0: H0:meaw=75
Ha='sample mean different from population mean , the exam score significantly differs from 75'    # Ha: Ha:meaw<>75
# As meaw<>75 and sample size is <30, so it is a 2 tail t test

# Step2 - Set the threshold
    # As alpha is not given then we can take default alpha
alpha=0.05

# step 3 - Colllect the evidences
population_mean=75 # meaw
#pop_std_dev=? #sigma_std_dev
sample_size=20 #n
#sample_mean=?

# Step 4 - find p and z value
# t-test

t_stat,p_value=stats.ttest_1samp(sample_Scores,population_mean,alternative='two-sided' )
    #,alternative='two-sided' -- optional as bydefault is two sided only
    # alternative='less'  -- left tail test
    # # alternative='greater'  -- right tail test
    # formula for 1sample ttest
print(f" tstat : {t_stat:,.3f}")
print(f" pvalue : {p_value}")

# Step 5 - conclusion compare p and alpa
if p_value<alpha:
    print(f' reject the null hypothesis : {Ha} ' )
else:
    print(f' failed to reject null hypothesis :  {H0} ')
```

## 18.3 one proportion z-test

#### Ex - 18.3.1 Custome rsatifaction rate
A company claims that 80% of its cusotomers are satisfied with their services. A survey of 150 customers finds that 110 are satisfied. We test whether 
the satisfactin rate is different from 80%.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

#ttest
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel
#ztest
#import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest


# Step1 - set up the null hypothesis
H0='80% of customers are satisfied'    # H0: p:meaw=80
Ha='80% of customers are not satisfied'# Ha: p:meaw<>80
# As input dataset is catogorical, meaw<>80 and sample size is >30, so it is a 1 sample - two tail - t test

# Step2 - Set the threshold
    # As alpha is not given then we can take default alpha
alpha=0.05

# step 3 - Colllect the evidences
population_mean=80 # meaw
#pop_std_dev=? #sigma_std_dev
claimed_proportion=0.80
sample_size= 150
sample=110

# Step 4 - find p and z value
# 1 prop 2 tailed test

# alternative='two-sided' -- optional as bydefault is two sided only
# alternative='smaller'  -- left tail test
# alternative='larger'  -- right tail test

z_stat,p_value=proportions_ztest(count=sample,nobs=sample_size,value=claimed_proportion,alternative='two-sided')

print(f" z-stastics : {z_stat:,.3f}")
print(f" pvalue : {p_value}")

# Step 5 - conclusion compare p and alpa
if p_value<alpha:
    print(f' reject the null hypothesis : {Ha} ' )
else:
    print(f' failed to reject null hypothesis :  {H0} ')
```

A school wants to check if a training programme improvess students test scores. We compare scpres before and after for the same students.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

#ttest
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel
#ztest
import statsmodels.api as sm

before=[68,66,62,67,73,75,65,63,70,70]
after=[75,86,76,77,81,75,80,76,71,61]

# Step1 - set up the null hypothesis
H0='training programme does not improves students score significantly'   # H0: p:meaw1>=meaw2
Ha='training programme improves students score significantly'# Ha: p:meaw1<meaw1
# As input dataset is numerical , 2 samples, p:meaw2>meaw1 and sample size is <30 i.e.10, so it is a 2 sample piared - left tail - t test

# Step2 - Set the threshold
    # As alpha is not given then we can take default alpha
alpha=0.05

# step 3 - Colllect the evidences
#before=[68,66,62,67,73,75,65,63,70,70]
#after=[75,86,76,77,81,75,80,76,71,61]

# Step 4 - find p and z value
# 1 prop 2 tailed test

t_stat,p_value=stats.ttest_rel(before,after,alternative='less')

print(f" t-stastics : {t_stat:,.3f}")
print(f" pvalue : {p_value}")

# Step 5 - conclusion compare p and alpa
if p_value<alpha:
    print(f' reject the null hypothesis : {Ha} ' )
else:
    print(f' failed to reject null hypothesis :  {H0} ')
```
