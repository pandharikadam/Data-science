- By looking at the above diagram we can say that there are very few outliers are available

- lets find out the outliers by IQR

```python
q1=cardio_df_notnull['LB'].quantile(0.25)
q3=cardio_df_notnull['LB'].quantile(0.75)
print(q1,q3)
```

```python
IQR=q3-q1
print(IQR)
```

```python
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
print(lower_bound,upper_bound)
```

```python
outliers=cardio_df_notnull['LB'][(cardio_df_notnull['LB']<lower_bound) | (cardio_df_notnull['LB']>upper_bound)]
print(outliers)
```

- As data is normally distributed we can not use IQR method for outlier removal, better use zscore method

```python
outliers.count()
```

```python
from scipy import stats
```

```python
filter=pd.DataFrame(stats.zscore(cardio_df_notnull["LB"]))>3
print(filter)
filter.value_counts()
```
