## 33.2.1 OHE encoding - One hot encoding

```python
iris_ohe=iris.copy()
iris_ohe.head()
```

```python
ohe_enc=OneHotEncoder()
#ohe_enc.fit_transform([iris_label['species']]) # it is not possible to see output directly as it is parse matridx, so convert it into array and then arrat
pd.DataFrame(ohe_enc.fit_transform(iris_ohe[['species']]).toarray())
```

```python
iris_final_ohe=iris_ohe.join(pd.DataFrame(ohe_enc.fit_transform(iris_ohe[['species']]).toarray()))
iris_final_ohe
```

```python
iris_final_ohe.drop('species',axis=1,inplace=True)
```

```python
iris_final_ohe.head()
```

```python
iris_final_ohe.rename(columns={0:'species_setos',1:'species_versicolor',2:'species_virgina'},inplace=True)
```

# 33.3 Isolation forest

```python
from sklearn.ensemble import IsolationForest
```

```python
iris.head(3)
```

```python
iris_features=iris.iloc[:,0:4]
```

```python
iso=IsolationForest(contamination=0.02)
iso.fit(iris_features)
```

```python
iso.predict(iris_features)
```

```python
iris_features['anamoly']=iso.predict(iris_features)
```
