### Calculate probability of continuoues random variable

```python
#E1 Find the probabilty of getting the marks <=680,mean=711,std=29
```

```python
p_less_680=stat.norm.cdf(680,loc=711,scale=29) # thi sformula gives the zscore(standardised value) for 680
np.round(p_less_680*100,2)
# loc - mean
# scale - std deviation
#14.25 percent of probability of getting marks less than 680
```

```python
#E2 Find out the probabilty of getting the marks >720,mean=711,std=29
# As we can not find the right side area(grater than) of any vaule, first we need to find out less than arean and minus it with 1
```

```python
p_less_720=stat.norm.cdf(720,loc=711,scale=29)
print(p_less_720)
p_greater_720=1-p_less_720
print(np.round(p_greater_720*100,2))
```

```python
import scipy
p_less_720=scipy.stats.norm.cdf(720,loc=711,scale=29)
print(p_less_720)
p_greater_720=1-p_less_720
print(np.round(p_greater_720*100,2))
```

```python
#E3 Find out the probabilty of getting the between 680<= marks <= 720,mean=711,std=29
```

```python
probability_between_680_720=p_less_720-p_less_680
print(np.round(probability_between_680_720*100,2))
#p_less_680+probability_between_680_720+p_greater_720 should be equal to 100 as area under the curve is always 100
```

## 15.2 Central limit theorem

# Confidence interval - using z distruition

```python

c_int_z95=stat.norm.interval(0.95,loc=1990,scale=2500/np.sqrt(140))
np.round(c_int_z95)
```

```python
c_int_z90=stat.norm.interval(0.90,loc=1990,scale=2500/np.sqrt(140))# loc - sample mean
np.round(c_int_z90)
```

```python
c_tax_z90=stat.norm.interval(0.90,loc=23.6,scale=7/np.sqrt(100))
np.round(c_tax_z90)
```

# Confidence interval - using t distruition

```python
c_int_t95=stat.t.interval(0.95,df=139,loc=1990,scale=2833/np.sqrt(140))
np.round(c_int_t95)#df=n-1=140-1=139
# 2833 - population std dev
# loc - 1990 - sample mean
```

```python
c_tax_t90=stat.t.interval(0.90,df=99,loc=23.6,scale=7/np.sqrt(140))
print(c_tax_t90)
```
