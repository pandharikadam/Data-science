```python
# As it is talking about wight(number) and sample size is more than 30 and the n of samples are only one.
# So the test conducted would be "1 sample z test"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
#ttest
from scipy.stats import ttest_1samp,ttest_ind,ttest_rel
#ztest
import statsmodels.api as sn



# Step1 - set up the null hypothesis
H0='Avg weight of individual is 70kg'    # H0:Avg weight of individual is 70kg - H0:meaw=70
Ha='Avg weight of individual is not equal to 70kg'    # Ha:Avg weight of individual is not equal to 70kg - Ha:meaw<>70
# As meaw<>70 so it is a 2 tail test

# Step2 - Set the threshold
    # As alpha is not given then we can take default alpha
alpha=0.05

# step 3 - Colllect the evidences
population_mean=70 # meaw
pop_std_dev=3.5 #sigma_std_dev
sample_size=40 #n
sample_mean=67.9

# Step 4 - find p and z value
# Z-test

Z_score_weight = (sample_mean-population_mean)/(pop_std_dev/np.sqrt(sample_size))
    # We dont have z-test readily  availabe so we have built our own procees to conduct z-test
print(f" zscore : {Z_score_weight:,.3f}")
    #round(Z_score_weight,2)
p_value_weight = 2*stats.norm.cdf(Z_score_weight) # as it is a 2 tail test, always we compare p=2alpha, so multiplied by 2 in formula
print(f" pvalue : {p_value_weight}")

# Step 5 - conclusion compare p and alpa
if p_value_weight<alpha:
    print(f' reject the null hypothesis : after new diet : {Ha} : diet changes the averege weight of individuals')
else:
    print(f' fail to reject null hypothesis : after new diet : {H0} : diet not changes the averege weight of individuals')
```
